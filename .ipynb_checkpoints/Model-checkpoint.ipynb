```python
def remove_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Model'):
        m = np.mean(subdf.price)
        st = np.std(subdf.price)
        reduced_df = subdf[(subdf.price>(m-st)) & (subdf.price<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_outliers(df6)
df7.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.linear_model import Lasso
ls_clf = Lasso(tol=0.1)
ls_clf.fit(X_train,y_train)
ls_clf.score(X_test,y_test)
```

```python
from sklearn.linear_model import Ridge
rg_clf = Ridge()
rg_clf.fit(X_train,y_train)
rg_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'tol': [0.1, 1],
                'selection': ['random', 'cyclic']
            }
        },
        'ridge': {
            'model': Ridge(),
            'params': {
                'alpha': [1,2],
                'tol': [0.001, 0.0001]
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(brand,model,year,mileage,fuel,transmission,condition,capacity):    
    brand_index = np.where(X.columns==brand)[0][0]
    model_index = np.where(X.columns==model)[0][0]
    capacity_index = np.where(X.columns==capacity)[0][0]
    transmission_index = np.where(X.columns==transmission)[0][0]
    fuel_index = np.where(X.columns==fuel)[0][0]
    condition_index = np.where(X.columns==condition)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = mileage
    x[308]=2022-year
    if brand_index >= 0:
        x[brand_index] = 1
        
    if model_index >= 0:
        x[model_index] = 1
        
    if capacity_index >= 0:
        x[capacity_index] = 1
        
    if transmission_index >= 0:
        x[transmission_index] = 1
        
    if fuel_index >= 0:
        x[fuel_index] = 1
        
    if condition_index >= 0:
        x[condition_index] = 1

    return rg_clf.predict([x])[0]
```
