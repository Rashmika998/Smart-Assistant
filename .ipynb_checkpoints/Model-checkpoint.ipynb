```python
def remove_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Model'):
        m = np.mean(subdf.price)
        st = np.std(subdf.price)
        reduced_df = subdf[(subdf.price>(m-st)) & (subdf.price<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_outliers(df6)
df7.shape
```

```python
df8=df7.copy()
dummies=pd.get_dummies(df8.Transmission)
```

```python
dummies_fuel = pd.get_dummies(df10.Fuel)
df11 = pd.concat([df10,dummies_fuel],axis='columns')
df11.head()
```

```python
df12 = df11.drop(['Condition','Transmission','Fuel'],axis='columns')
df12.head()
```

```python
dummies_model = pd.get_dummies(df12.Model)
df13 = pd.concat([df12,dummies_model],axis='columns')

dummies_brand = pd.get_dummies(df13.Brand)
df14 = pd.concat([df13,dummies_brand],axis='columns')
df14.head()
```

```python
df15 = df14.drop(['Model','Brand'],axis='columns')
df15.head()
```

```python
df15.shape
```

```python
dummies_year = pd.get_dummies(df15.Year)
df16 = pd.concat([df15,dummies_year],axis='columns')

dummies_cap = pd.get_dummies(df16.capacity)
df17 = pd.concat([df16,dummies_cap],axis='columns')

df18 = df17.drop(['capacity'],axis='columns')
df18.head()
```

```python
X = df18.drop(['price'],axis='columns')
X.head(3)
```

```python
X.shape
```

```python
y = df18.price
y.head(3)
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```
